```python
from chainer import reporter
from chainer.training import util


class MergedTrigger(object):

    def __init__(self, *triggers):
        self._triggers = triggers

    def __call__(self, trainer):
        ret = False
        for trigger in self._triggers:
            # Do not break the loop even if ret = True because some triggers
            # expect to be called after every iteration
            ret = (trigger(trainer) or ret)
        return ret
                

class ThresholdTrigger(object):

    """Trigger invoked when specific value becomes lower than the threshold value.

    Args:
        key (str): Key of value.
        thresh: Threshold value.
        trigger: Trigger that decides the comparison interval between current
            best value and new value. This must be a tuple in the form of
            ``<int>, 'epoch'`` or ``<int>, 'iteration'`` which is passed to
            :class:`~chainer.training.triggers.IntervalTrigger`.

    """

    def __init__(self, key, thresh, trigger=(1, 'epoch')):
        self._key = key
        self._thresh = thresh
        self._interval_trigger = util.get_trigger(trigger)
        self._init_summary()

    def __call__(self, trainer):
        """Decides whether the extension should be called on this iteration.

        Args:
            trainer (~chainer.training.Trainer): Trainer object that this
                trigger is associated with. The ``observation`` of this trainer
                is used to determine if the trigger should fire.

        Returns:
            bool: ``True`` if the corresponding extension should be invoked in
                this iteration.

        """

        observation = trainer.observation
        summary = self._summary
        key = self._key
        if key in observation:
            summary.add({key: observation[key]})

        if not self._interval_trigger(trainer):
            return False

        stats = summary.compute_mean()
        value = float(stats[key])  # copy to CPU
        self._init_summary()

        if self._thresh > value:
            return True
        return False

    def _init_summary(self):
        self._summary = reporter.DictSummary()
```

```python
from sklearn.exceptions import NotFittedError
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import chainer
from chainer import cuda
from chainer import functions as F
from chainer import links as L
from chainer import optimizers
from chainer.training import extensions


class MLP(chainer.Chain):
    def __init__(self, n_out):
        super(MLP, self).__init__()
        with self.init_scope():
            # the size of the inputs to each layer will be inferred
            self.l1 = L.Linear(None, n_out)  # n_in -> n_units

    def __call__(self, x):
        return self.l1(x)


class SVMClassifier(chainer.Chain):
    def __init__(self, predictor, penalty):
        super(SVMClassifier, self).__init__()
        with self.init_scope():
            self.predictor = predictor
        self.penalty = penalty

    def __call__(self, x, t):
        y = self.predictor(x)
        loss = F.hinge(y, t, norm=self.penalty)
        accuracy = F.accuracy(y, t)
        chainer.report({'loss': loss, 'accuracy': accuracy}, self)
        return loss


class LogitClassifier(chainer.Chain):
    def __init__(self, predictor):
        super(LogitClassifier, self).__init__()
        with self.init_scope():
            self.predictor = predictor

    def __call__(self, x, t):
        y = self.predictor(x)
        loss = F.softmax_cross_entropy(y, t)
        accuracy = F.accuracy(y, t)
        chainer.report({'loss': loss / len(x), 'accuracy': accuracy}, self)
        return loss


class SklearnSVCInterface(object):
    def __init__(self, C=1.0, verbose=0, max_iter=1000, batchsize=32, tol=1e-2):
        self._model = None
        self._scaler = None
        self._whitener = None
        self._verbose = verbose
        self._C = C
        self._max_iter = max_iter
        self._batchsize = batchsize
        self._tol = tol

    def _init_model(self, X, y):
        raise NotImplementedError()

    def fit(self, X, y):
        self._model = self._init_model(X, y)
        # Apply scaling and whitening because NN is prone to variance
        self._scaler = MinMaxScaler().fit(X)
        X = self._scaler.transform(X)
        self._whitener = PCA(whiten=True).fit(X)
        X = self._whitener.transform(X)
        optimizer = optimizers.MomentumSGD()
        optimizer.setup(self._model)
        # To keep total loss stable, I convert original SVM loss
        # `1/2 * decay + C * hinge_loss` to `1/2C * decay + 1 * hinge_loss`
        optimizer.add_hook(chainer.optimizer.WeightDecay(0.5 / float(C)))
        train = chainer.datasets.TupleDataset(X.astype(np.float32), y.astype(np.int32))
        train_iter = chainer.iterators.SerialIterator(train, self._batchsize)

        updater = chainer.training.StandardUpdater(train_iter, optimizer)
        # Stop after max_iter or when loss is lower than tol value
        stop_trigger = MergedTrigger(
            chainer.training.IntervalTrigger(self._max_iter, 'epoch'),
            ThresholdTrigger('main/loss', self._tol))

        trainer = chainer.training.Trainer(updater, stop_trigger, out='result')
        trainer.extend(extensions.ExponentialShift("lr", 0.995))
        trainer.extend(extensions.LogReport())
        if self._verbose >= 1:
            trainer.extend(extensions.PrintReport(['epoch', 'main/accuracy', 'main/loss', 'lr']))
            trainer.extend(extensions.observe_lr())
        trainer.run()
        return self

    def fit_transform(self, X, y):
        self.fit(X, y)
        return self.predict(X)

    def predict(self, X):
        X = self._scaler.transform(X)
        X = self._whitener.transform(X)
        return np.argmax(self.decision_function(X), axis=1)

    def decision_function(self, X):
        if self._model is None:
            raise NotFittedError("This %(name)s instance is not fitted "
                                 "yet" % {'name': type(self).__name__})
        X = chainer.Variable(X.astype(np.float32))
        y = self._model.predictor(X)
        return cuda.to_cpu(y.data)


class SVM(SklearnSVCInterface):
    def __init__(self, penalty='l2', C=1.0, verbose=0, max_iter=1000, batchsize=32,
                 tol=1e-2):
        super(SVM, self).__init__(C=C, verbose=verbose, max_iter=max_iter,
                                  batchsize=batchsize, tol=tol)
        self._penalty = penalty.upper()

    def _init_model(self, X, y):
        n_class = np.max(y.data) + 1
        return SVMClassifier(MLP(n_class), self._penalty)


class LogisticRegression(SklearnSVCInterface):
    def _init_model(self, X, y):
        n_class = np.max(y.data) + 1
        return LogitClassifier(MLP(n_class))
```

```python
import numpy as np
import matplotlib.pyplot as plt


def draw(X, y, classifiers, hplot, wplot, h=.02, spacing=1.):
    # h: step size in the mesh
    # classifiers (dict)

    matplotlib.rcParams.update({'font.size': 6})
    # create a mesh to plot in
    x_min, x_max = X[:, 0].min() - spacing, X[:, 0].max() + spacing
    y_min, y_max = X[:, 1].min() - spacing, X[:, 1].max() + spacing
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    for i, (title, clf) in enumerate(classifiers.iteritems()):
        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, m_max]x[y_min, y_max].
        plt.subplot(hplot, wplot, i + 1)
        plt.subplots_adjust(wspace=0.1, hspace=0.2)
        clf.fit(X, y)
        pred = clf.predict(X)
        accu = np.sum(pred == y) / float(len(X))
        title += " (Acc={:0.2f})".format(accu)
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.5,
                     vmin=0, vmax=max(y))

        # Plot also the training points
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.xticks(())
        plt.yticks(())
        plt.title(title)

    plt.show()
```

## Against artificial data 1: blob

```python
from collections import OrderedDict
from sklearn.datasets.samples_generator import make_blobs
from sklearn import svm

C = 1.0  # SVM regularization parameter
# title for the plots
classifiers = OrderedDict((
    ('SVC', svm.SVC(kernel='linear', C=C)),
    ('LinearSVC', svm.LinearSVC(C=C)),
    ('NN SVC', SVM(C=C, verbose=0)),
    ('NN Logistic Regression', LogisticRegression(C=C, verbose=0))
))

# Gaussian blob
X, y = make_blobs(n_samples=300, centers=2, n_features=2,
                  random_state=0)

draw(X, y, classifiers, 2, 2, spacing=0.2)
```

## Reference: comparison against [sklearn.linear_model.SGDClassifier](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html#sklearn-linear-model-sgdclassifier)

Compare against sklearn's implementation of perceptron. I will leave parameters to default to see if there is any "magics" in it's parameter tuning.

```python
from collections import OrderedDict
from sklearn.linear_model import SGDClassifier

np.random.seed(1)
X = np.vstack((np.random.normal((-0.3, -0.3), 0.35, (290, 2)),
               np.random.normal((0.4, 0.4), 0.05, (5, 2)),
               np.random.normal((0.4, 0.4), 0.1, (5, 2))))
y = np.array([0] * 290 + [1] * 10, dtype=int)
perm = np.random.permutation(len(X))
X = X[perm]
y = y[perm]

C = 20.0  # SVM regularization parameter
classifiers = OrderedDict((
    ('SVC C=%0.1f' % C, svm.SVC(kernel='linear', C=C)),
    ('NN SVC C=%0.1f' % C, SVM(C=C, verbose=0)),
    ('SGDClassifier (SVM)', SGDClassifier(loss='squared_hinge')),
    ('SGDClassifier (Perceptron)', SGDClassifier(loss='log')),
))

draw(X, y, classifiers, 2, 2, spacing=0.2)
```
